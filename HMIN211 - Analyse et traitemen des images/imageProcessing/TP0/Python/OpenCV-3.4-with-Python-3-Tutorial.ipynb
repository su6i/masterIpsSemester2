OpenCV module

```python
# Read image
imageCv2 = cv2.imread("../../images/jpg/lena4.jpg")
# gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# cv2.imwrite("gray_panda.jpg", gray_image)
lenaFlipped = cv2.flip(imageCv2,0)
cv2.imshow("CV2 flipped",lenaFlipped)

# Close windows and waitKey to hold image before closing
cv2.waitKey(0)
cv2.destroyAllWindows()
```
